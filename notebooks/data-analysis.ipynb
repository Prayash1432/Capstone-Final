```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

## Data Preprocessing

```python
dataset=pd.read_csv('../data/Final_dataset_CS.csv',header=1)
print(dataset.shape)
dataset = dataset.drop(dataset.columns[[0,1,2]], axis=1)
dataset.head()
```

```python
dataset['Water (kg/m3)'] = pd.to_numeric(dataset['Water (kg/m3)'], errors='coerce')
dataset.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder()
dataset['Condition'] = le.fit_transform(dataset['Condition'])
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
dataset = dataset.copy()
dataset[:] = imputer.fit_transform(dataset)
dataset.head()
```

```python
(dataset.isnull().mean())*100
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm')
plt.savefig('correlation_plot.png', dpi=300)
plt.show()
```

## Feature Engineering 

```python
df = pd.read_csv('../data/new_ds.csv', header=0)
df.head()
```

```python
# total binders

original_features = df.columns.to_list()
total_binders = ['GGBS_kg_m3', 'FlyAsh_kg_m3', 'SF_kg_m3']
df['Total_Binders'] = df['GGBS_kg_m3'] + df['FlyAsh_kg_m3'] + df['SF_kg_m3']
df.head()
```

```python
# total SiO2

original_features = df.columns.to_list()
total_sio2 = ['GGBS_SiO2_%', 'FA_SiO2_%', 'SF_SiO2_%']
df['Total_SiO2'] = ((df['GGBS_kg_m3'] * df['GGBS_SiO2_%']) + (df['FlyAsh_kg_m3'] * df['FA_SiO2_%']) + (df['SF_kg_m3'] * df['SF_SiO2_%'])) / df['Total_Binders']
df.head()
```

```python
# total CaO

original_features = df.columns.to_list()
total_cao = ['GGBS_CaO_%', 'FA_CaO_%', 'SF_CaO_%']
df['Total_CaO'] = ((df['GGBS_kg_m3'] * df['GGBS_CaO_%']) + (df['FlyAsh_kg_m3'] * df['FA_CaO_%']) + (df['SF_kg_m3'] * df['SF_CaO_%'])) / df['Total_Binders']
df.head()
```

```python
# total Al2O3

original_features = df.columns.to_list()
total_Al2O3 = ['GGBS_Al2O3_%', 'FA_Al2O3_%', 'SF_Al2O3_%']
df['Total_Al2O3'] = ((df['GGBS_kg_m3'] * df['GGBS_Al2O3_%']) + (df['FlyAsh_kg_m3'] * df['FA_Al2O3_%']) + (df['SF_kg_m3'] * df['SF_Al2O3_%'])) / df['Total_Binders']
df.head()
```

```python
# ratios
df['SiO2 / Al2O3'] = df['Total_SiO2'] / df['Total_Al2O3']
df['CaO / Al2O3'] = df['Total_CaO'] / df['Total_Al2O3']
df['CaO / SiO2'] = df['Total_CaO'] / df['Total_SiO2']
```
