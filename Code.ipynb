```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv('Final_dataset_CS.csv',header=1)
print(dataset.shape)
dataset.head()
```

```python
dataset = dataset.drop(dataset.columns[[0,1,2]], axis=1)
```

```python
dataset.head()
```

```python
plt.scatter(dataset["GGBS_SiO2_%"], dataset["CS 28 day (Mpa)"])

plt.xlabel("GGBS_SiO2_%")
plt.ylabel("Compressive Strength")
plt.title("GGBS_SiO2_% vs Strength")

plt.show()
```

```python
plt.scatter(dataset['SiO2 to Na2O (silicate modulus)'],dataset['CS 28 day (Mpa)'])
plt.xlabel("SiO2 to Na2O (silicate modulus)")
plt.ylabel("Compressive strength")
plt.show()
```
