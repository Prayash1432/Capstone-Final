```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    "font.size": 12,
    "font.family": "serif"
})

# Model names
models = ['Linear Reg', 'Random Forest', 'XGBoost', 'MLP']
cindex_values = [0.69, 0.87, 0.91, 0.98,]

# Colors
colors = ["#1977ba", "#d66808", "#0aa90a", "#7d22d2", "#a7a2ab"]

fig, ax = plt.subplots(figsize=(4.6, 3.5), dpi=300)

x = np.arange(len(models))

bars = ax.bar(x, cindex_values, color=colors, linewidth=0.8)

# Add value labels
for i, val in enumerate(cindex_values):
    ax.text(i, val + 0.005, f"{val:.2f}", ha='center', fontsize=10, fontweight='bold')

# Remove x-axis labels
ax.set_xticks([])

ax.set_ylim(0.6, 1.1)

# Axis label
ax.set_ylabel("$R^2$", fontsize=12)

# Grid
ax.grid(axis='y', linestyle='--', alpha=0.4)
ax.set_axisbelow(True)

# Legend
ax.legend(bars, models, frameon=False, loc='upper left')

# Remove unnecessary spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig("barplot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    "font.size": 12,
    "font.family": "serif"
})

# Model names
models = ['Linear Reg', 'Random Forest', 'XGBoost', 'MLP']
cindex_values = [17.95, 11.29, 9.63, 3.28]

# Colors
colors = ["#1B9E77", "#D95F02", "#7570B3", "#E7298A", "#66A61E"]

fig, ax = plt.subplots(figsize=(4.6, 3.5), dpi=300)

x = np.arange(len(models))

bars = ax.bar(x, cindex_values, color=colors, linewidth=0.8)

# Add value labels
for i, val in enumerate(cindex_values):
    ax.text(i, val + 0.1, f"{val:.2f}", ha='center', fontsize=10, fontweight='bold')

# Remove x-axis labels
ax.set_xticks([])

ax.set_ylim(1.0, 20.0)

# Axis label
ax.set_ylabel("Root Mean Sqaure Error (RMSE)", fontsize=12)

# Grid
ax.grid(axis='y', linestyle='--', alpha=0.4)
ax.set_axisbelow(True)

# Legend
ax.legend(bars, models, frameon=False, loc='upper right')

# Remove unnecessary spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig("rmse_barplot.png", dpi=300, bbox_inches='tight')
plt.show()
```
