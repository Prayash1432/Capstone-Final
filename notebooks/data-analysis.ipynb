```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

## Data Preprocessing

```python
dataset=pd.read_csv('Final_dataset_CS.csv',header=1)
print(dataset.shape)
dataset = dataset.drop(dataset.columns[[0,1,2]], axis=1)
dataset.head()
```

```python
dataset['Water (kg/m3)'] = pd.to_numeric(dataset['Water (kg/m3)'], errors='coerce')
dataset.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder()
dataset['Condition'] = le.fit_transform(dataset['Condition'])
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
dataset = dataset.copy()
dataset[:] = imputer.fit_transform(dataset)
dataset.head()
```

```python
(dataset.isnull().mean())*100
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm')
plt.savefig('correlation_plot.png', dpi=300)
plt.close()
```
